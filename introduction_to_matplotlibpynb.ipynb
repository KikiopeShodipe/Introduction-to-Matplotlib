# Student's Average Marks
* WE aim to find out the average performance of students ina a class and visulaize their overall performance

```python
#import the libaries
import pandas as pd #for data manipulation
import matplotlib.pyplot as plt#for data visualization
```

#Data loading
* Create a dataframe with students names and marks.


```python
student_names=["Susan","John","Mike","Kelly","Mary"]
#student marks
student_marks=[23,46,67,78,90]
```

# Calculate the average score
* Calculate what each student scored in items of percentages and averages

```python
#create an empty list to store percentages scores
marks_percentage=[]
#loop
for x in student_marks:
  #calculate the percentage
  result=(x/200)*100
  #add the percentage scores to the list
  marks_percentage.append(result)
```

# Representation of students grades
* General overview of the class performance

```python
#linechart to show student performance
plt.plot(student_names,marks_percentage)
#title
plt.title("Overall class performance")
#add x and y labels
plt.xlabel("Students")
plt.ylabel("Percentage")
#show the chart
plt.show()
```

```python
#barchart to show student performance
plt.bar(student_names,marks_percentage)
#title
plt.title("Overall class performance")
#add x and y labels
plt.xlabel("Students")
plt.ylabel("Percentage")
#show the chart
plt.show()
```
